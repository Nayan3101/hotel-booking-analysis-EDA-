### Variables Description

Based on the provided table, here is the variable description for each column:
1. Hotel : type of hotel
2. is_canceled: Whether the booking was canceled (1) or not canceled (0).

4. arrival_date_year: Year of arrival date.
5. arrival_date_month : month of arrival
6. arrival_date_week_number: Week number of the arrival date.
7. arrival_date_day_of_month: Day of the month of the arrival date.
8. stays_in_weekend_nights: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.
9. stays_in_week_nights: Number of weeknights (Monday to Friday) the guest stayed or booked to stay at the hotel.
10. adults: Number of adults included in the booking.
11. children: Number of children included in the booking.
12. babies: Number of babies included in the booking.
13. meal : type of meal provided
14. country : Guest country name
15. is_repeated_guest: Whether the booking was made by a repeated guest (1) or not (0).
16. previous_cancellations: Number of previous booking cancellations by the same guest.
17. previous_bookings_not_canceled: Number of previous bookings that were not canceled by the same guest.
18. booking_changes: Number of changes made to the booking.
19. agent: ID of the travel agency that made the booking.
20. company: ID of the company or entity that made the booking.
21. days_in_waiting_list: Number of days the booking was in the waiting list before it was confirmed.
22. adr: Average daily rate
23. required_car_parking_spaces: Number of car parking spaces required by the guest.
24. total_of_special_requests: Number of special requests made by the guest.
25. reservation_status : weather the guest have 'Check-Out','Canceled'or 'No-Show' at the hotel
26. reservation_status_date : on which date guest have Check-Out','Canceled'or 'No-Show' at the hotel
